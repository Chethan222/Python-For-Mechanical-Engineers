#### Aim:-To create python program to create the animation of pendulum motion

#### Equation used:- The ordinary differential equation which describes the pendulum motion is given by,

$\frac{d^2theta}{dt^2}+\frac{b}{m}*\frac{d \theta}{dt}+\frac{g}{l}*sin\theta = 0$

$\theta $:Angle between the vertical axis and the pendulum thread.

m: Mass of the bob.

l: Length of the thread.

g: Acceleration due to gravity.

b: Damping coefficient.

#### Objectives:-

     * To  create python program which gives the solution for the ode which describes the pendulum motion .

     *To create the animation of the pendulum moment and hence analyze its motion.     

#### Procedure:-

    The above given equation describes the pendulum motion. The necessary initial conditions are given to find out the solution for this ODE. At the beginning of the program necessary modules like math, matplotlib. pyplot, numpy, scipy are imported. Each modules are required to perform certain calculations. In this program one of the function of scipy module is used for find out the solution for the ode which describes the pendulum motion. The variables for length of the pendulum thread, mass of the bob, damping coefficient,gravity are assigned with given values. The linspace function of numpy is used to generate the different time points between 0 to 20 seconds. The ODE is solved by the odeint function of scipy module which takes function as argument that generates the ode . So a function is created to generate the ODE for the pendulum motion which takes mass of the bob, thread length, damping coefficient, time points, initial angular displacement and angular velocity as the arguments and generates  and returns ODE. The  above given ODE is generated by combining three terms of the equation. The initial conditions and the ODE is passed to the odeint function which generates the solution for the given time point. All possible values of angles are stored in the thetas list. Since , for each time point two solution for  angles are possible. So they are separated by using forloop and store them in two different lists. To generate the png file describing the pendulum position , the logic is used in such a way that one point or end of the thread is fixed to a support at d distance more than the sum of length of the string and bob diameter. The another point is computed by the algebra using corresponding value of angle between vertical reference line and the thread line. Each positions are plotted and the bob is created by using marker attribute of plot function using appropriate marker size.  Each png files are named by using string formatting method. When all the png files are generated , an animation is created with the help of imagemagic software.

#### Conclusion:-

   *Using the ODE we can easily analyze the motion of the pendulum.
