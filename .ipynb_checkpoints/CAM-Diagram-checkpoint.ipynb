```python
#Cam with SHM

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Specify number of divisions
n = 50

#Camshaft displacement
h = 25

#Step anglefor SHM
stepAngle = 360/n 

#Specify motion parameters(in degrees)
out = 60
dwell = 90
return_ = 120
offset = 0


```

```python

angles = np.linspace(0,180,n)
angles1 = np.linspace(0,out,n)
angles2 = np.linspace(out,out+dwell,n)
angles3 = np.linspace(out+dwell,out+dwell+return_,n)
angles4 = np.linspace(out+dwell+return_,360,n)

y_1 = np.array([(h/2 + h/2*math.cos(math.radians(angle))) for angle in angles])[::-1]
y_2 = h * np.ones_like(angles2)
y_3 = y_1[::-1]
y_4 = np.zeros_like(angles4)


```

```python
x_ax = np.concatenate((angles1,angles2,angles3,angles4))
y_ax = np.concatenate((y_1,y_2,y_3,y_4))
h_ = h*np.ones_like(x_ax)

#CAM-Profile
x_ = (h_+y_ax)*np.cos(np.radians(x_ax))
y_ = (h_+y_ax)*np.sin(np.radians(x_ax))

#Basic circle
x,y = h*np.cos(np.radians(x_ax)),h*np.sin(np.radians(x_ax))

 #Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(1)

plt.grid(True)
axis.set_xlabel('Theta')
axis.set_ylabel('Displacement')
axis.plot(x_ax,y_ax)
```

```python

plt.gca().set_aspect('equal', adjustable='box')

#Plotting circle
plt.plot(x,y)
plt.plot(x_,y_)
plt.show()
```
