```python
#Importing necessary modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading data
data = pd.read_csv('data.csv',index_col=0)
constants = pd.read_csv('CONSTANTS.csv',index_col=0)
xBar = data.XBar
R = data.R
```

```python

def computeControlLimits(x,xBar,R):
    global UCLx
    global LCLx
    global UCLr
    global LCLr
    global xBarBar
    global rBar
    global N
    
    N = xBar.shape[0]
    xBarBar = xBar.mean()
    rBar = R.mean()

    #Specification limits
    #specificationLimit = '171+-11'
    # tolerance = float(specificationLimit.split('+-')[1])
    # basicSize = float(specificationLimit.split('+-')[0])
    # USL = basicSize + tolerance
    # LSL = basicSize - tolerance

    #Parameters from table
    D4 =  constants.iloc[x-2]['D4']
    D3 =  constants.iloc[x-2]['D3']
    A2 =  constants.iloc[x-2]['A2']
    d2 =  constants.iloc[x-2]['d2']
    
    #Control limits
    UCLx = xBarBar + A2*rBar
    LCLx = xBarBar - A2*rBar

    UCLr = D4*rBar
    LCLr = D3*rBar

def check(xBar):
    if((xBar < UCLx).all() and (xBar > LCLx).all()):
        print('Process is in control.')
    else:
        print('Process is not in control!')
        

def plotRChart(R):
    fig, axis = plt.subplots(1, 1)
 
    #R-Chart
    #Control plot
    axis.plot(range(1,N+1),R,label='Control Plot')

    #R-Bar Line
    axis.plot([0,N],[rBar,rBar],label='R-Bar')

    #Upper Control Line
    axis.plot([0,N],[UCLr,UCLr],label='UCLr')

    #Lower Control Line
    axis.plot([0,N],[LCLr,LCLr],label='LCLr')
    
    axis.set_xlabel('Sample No.')
    axis.set_ylabel('R')
    axis.set_title('R-Chart')
    axis.legend(loc='upper right')
    plt.show()

def plotXBarChart(xBar):
    figure, axis = plt.subplots(1, 1)
    #XBar-Chart
    #Control plot
    axis.plot(range(1,N+1),xBar,label='Control Plot')

    #R-Bar Line
    axis.plot([0,N],[xBarBar,xBarBar],label='X-BarBar')

    #Upper Control Line
    axis.plot([0,N],[UCLx,UCLx],label='UCLx')

    #Lower Control Line
    axis.plot([0,N],[LCLx,LCLx],label='LCLx')
    
    axis.set_xlabel('Sample No.')
    axis.set_ylabel('X-Bar')
    axis.set_title('X-Bar Chart')
    axis.legend(loc='upper right')
    plt.show()

def removeOutliers(data):
    newData = data[(data.XBar < UCLx) & (data.XBar > LCLx) & (data.R < UCLr) & (data.R > LCLr) ] 
    print('Removed %d elements.' % (data.shape[0]-newData.shape[0]))
    return newData

```

```python
#Sample parameters
x = 5

#Computing control limits
computeControlLimits(x,xBar,R)

#Checking process control
check(xBar)
```

```python

#Plotting control charts

#R-Chart
plotRChart(R)
```

```python
plotXBarChart(xBar)
```

```python
#Bringing process to control
newData =removeOutliers(data)
```

```python
newDataXBar = newData.XBar
newDataR = newData.R
newData
```

```python
#Computing control limits for new data
computeControlLimits(x,newDataXBar,newDataR)

#Checking process control
check(newDataXBar)
```

```python
plotRChart(newDataR)
```
